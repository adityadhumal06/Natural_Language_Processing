```python
# Lower

df['review'] = df['review'].str.lower()
```

```python
# Remove HTML Tags

import re
def remove_html_tags(text):
  pattern = re.compile('<.*?>')
  return pattern.sub(r'',text)
```

```python
df['review'].apply(remove_html_tags)
```

```python
# Remove URLs

import re
def remove_url(text):
  url_pattern = r'http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+'
  cleaned_text = re.sub(url_pattern,'',text)
  return cleaned_text
```

```python
df['review'].apply(remove_url)
```

```python
# Problem 2

# Find out the number of words in the entire corpus and also the total number of unique words(vocabulary) using just python

import re
# concatenate all the 'review' text into a single string
corpus = ' '.join(df['review'])

# tokenize the text into words using regular expressions
words = re.findall(r'\w+',corpus.lower())

# calculate the total number of words and the number of unique words
total_words = len(words)
unique_words = set(words)

print('total_words :',total_words)
print('unique_words :',unique_words)
```
